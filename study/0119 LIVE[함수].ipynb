map : (A, B)
A는 각각 적용할 함수
B는 반복할 것

```python
def my_magic_func(n):
    return n * 10

my_list = [1,2,3,4,5]

rlt = map(my_magic_func, my_list)
print(rlt)
print(list(rlt))
```

Filter (function, iterable)
iterable 모든 요소에 function 적용
True일때 filter obj로 반환

```python
def odd(n):
    return n % 2
numbers = [1,2,3]
result = filter(odd,numbers)
print(result, type(result))

print(list(result))
```

zip (*iterables)
iterable아니고 iterables!!
복수의 iterable 모아 튜플을 원소로 하는 obj를 반환

```python
name_list = ['신동민,' '서재현', '박영서', '이태성', '정예원', '이은석']
age_list = [17, 18, 22, 24, 25, 19]

for each in zip(name_list, age_list):
    print(each)
#튜플로 묶어준다
```

lambda 함수
lambda 매개변수 : 매개변수리턴값

```python
lambda x : x*x

def pow(x):
    return x*x
#같은 기능

#실행 ->
#(lambda x: x*x)()
print((lambda x: x*x)(4))
```

```python
def my_magic_func(n):
    return n * 10

map_obj = map(lambda n : n * 10, [1,2,3])
# my_magit_func 대신 람다로
rlt = list(map_obj)

print(rlt)
```

```python
def test(**kwargs):
    print(kwargs, type(kwargs))
    kwargs['name']
    return kwargs

test(name='aiden', age=21)
```

```python
def test(*args, **kwargs):
    print(kwargs, type(kwargs))
    kwargs['name']
    return kwargs

test(1,2,3,4, name='aiden', age=21, music='IU')
```
