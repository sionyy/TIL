```python
# selection sort 선택정렬

a=[4,7,1,3,5,2]

for i in range(len(a)-1):
    for j in range(i+1,len(a)):
        if a[i]>a[j]:
            a[i],a[j]=a[j],a[i]     # swap

for i in range(len(a)):
    print(a[i],end=' ')

for i in a:
  print(i,end='')
```

```python
# insert sort 삽입정렬

a=[4,7,1,3,5,2]

result=[]
for i in range(len(a)):
    result.append(a[i]) # 값 하나씩 내리기
    for j in range(i,0,-1): # 뒤에서 부터.. 앞으로 가면서
        if result[j-1]>result[j]: # 현재 vs 앞의값 비교
            result[j],result[j-1]=result[j-1],result[j]
        else:
            break
```

```python
# bubble sort 버블정렬

a = [8, 3, 12, 10, 1]


for i in range(len(a)-1,0,-1):  # 4 3 2 1
    for j in range (0,i): # i가 4일때 0123  i가 3일때 012...
        if a[j]>a[j+1]:
            a[j],a[j+1]=a[j+1],a[j]

print(a)
```
