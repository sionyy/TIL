클래스 / 객체 / 인스턴스 / 메서드

클래스 = 속성 / 메서드로 나뉨

    속성 = 클래스변수 / 인스턴스변수로 나뉨

```python
class calculator():
    numberOfCalcul=0            #속성 [클래스변수]
    
    def __init__(self):         #클래스 안에 함수 = 메서드
        self.result=0           #속성 [인스턴스 변수]        
        calculator.numberOfCalcul+=1
                                #__init__ = 객체를 만들자마자 바로 실행되는 것
                                #__init__ = 생성자함수(=constructor)
        

    def getsum(self,value):
        self.result+=value
        return self.result
    
cal1 = calculator()             #객체이자 calculator의 인스턴스(클래스와 객체의 관계)
cal2 = calculator()             #인스턴스변수란? 인스턴스 개인적 속성(attribute)

print(cal1.getsum(3))
print(cal1.getsum(4))
print(cal1.getsum(5))

print(cal2.getsum(6))
print(cal2.getsum(7))
print(cal2.getsum(1))


# 클래스 변수란?
# 클래스의 모든 인스턴스가 공유하는 값을 의미
print(cal1.numberOfCalcul)
print(cal2.numberOfCalcul)
print(calculator.numberOfCalcul)

#주의점 !!
# 클래스 변수값 변경시 항상
# 클래스.클래스변수 형식으로 변경해야함 [인스턴스.클래스변수로 변경 불가]
# 잘못된 예시[이렇게 하지 마세요]
cal1.numberOfCalcul=10
print(cal2.numberOfCalcul)

calculator.numberOfCalcul=20
print(cal1.numberOfCalcul)
print(cal2.numberOfCalcul) #클래스변수를 할당했지만 적용이 안되는 모습
```

```python
# 객체에 숫자 / 합과 차를 돌려주는 클래스 만들기
# 생성자 함수 없이 만들기

class plus_minus:
    def data(self, first, second):
        self.first = first
        self.second = second
        
    def plus(self):
        result = self.first + self.second
        return result
    
    def minus(self):
        result = self.first - self.second
        return result


a = plus_minus()
a.data(3,5)

b = plus_minus()
b.data(2,7)
print(a.first,b.second)

print(a.plus())
print(a.minus())

# 객체 a와 b는 각각 독립적
# data = 인스턴스 매서드
# 인스턴스 매서드의 첫번째 paremeter로 self를 사용 [관례적]
```

```python
# 생성자 함수 넣어서 만들기

class plus_minus:
    def __init__(self, first, second):      # 매직매서드
        self.first = first
        self.second = second
        
    def plus(self):
        result = self.first + self.second
        return result
    
    def minus(self):
        result = self.first - self.second
        return result
    

a = plus_minus(3,5)     # 인스턴스를 만듦과 동시에 3,5 할당
```

```python
class car:
    def __init__(self,name,price):
        self.name = name
        self.price = price

    def __add__(self, another):
        return self.price+another.price
    
    def __str__(self):
        return f'{self.name}의 가격은 {self.price}입니다.'
    
        
kia = car('k8', 300)
bmw = car('m5', 500)
hd = car('h2' , 100)
print(kia.price + bmw.price)

# 커스터마이즈 [.price안써도 되게하기]
print(kia+bmw+hd.price)


print(f'{kia.name}의 가격은 {kia.price}입니다.')

# __str__ 메서드 사용해서 print(kia)만 써도 되게 하기
print(kia)

# 인스턴스 삭제 : del
```

클래스 메소드 (데코레이터 사용하여 정의 / 호출시 첫번째 인자로 클래스 전달됨)
스태틱 메소드 (정적메소드) 클래스변수나 인스턴스 변수를 사용하지 않는 경우

인스턴스를 통하지 않고 클래스에서 바로 호출할 수 있는
클래스 메소드, 스태틱 매소드
@class method, @static method

```python
# 데코레이터
# 함수를 직접 수정하지 않고 함수의 기능을 수정하고자 할 때 사용

# 1. 데코레이터 사용 x

def call_name(name):
    print('반짝'*5)
    print(name)
    print('샤방'*5)
    
def call_age(age):
    print('반짝'*5)
    print(age)
    print('샤방'*5)
    
    
call_name('최민호')
call_age(39)

# 2. 데코레이터 사용 o

def deco(func): # 매개변수로 함수가 들어감
    
    def wrapping(value):
        print('반짝'*5)
        func(value)
        print('샤방'*5)
    return wrapping
    
@deco
def call_name(name):
    print(name)
    
@deco
def call_age(age):
    print(age)
    
call_name('최민호')
call_age(39)
```

```python
# 정적 매서드

class car:
    @staticmethod
    def add_price(one,another):
        print(one+another)
        
# 정적 메소드에는 @staticmethod를 붙임
# 그리고 self가 없다.
# add_price 함수의 매개변수를 넣는데 self 사용x

car.add_price(400, 800)

# 호출할 때 클래스에서 바로 매서드 호출
# 그러나 인스턴스가 있다면 인스턴스로도 static method 접근 가능
# 파이썬에서 가능(a7)

a7=car()
a7.add_price(500, 600)

# 인스턴스로 정적 매소드 호출은 잘 하지 않음
# 정적 메소드는 클래스의 인스턴스에 어떠한 변화도 일으키지 않는 함수라는 의미를 내포하고 암시할 떄 사용

```

```python
#클래스 매소드

# 몇명이 파이 만들었는지 확인하는 코드

class make_pie:
    cnt=0
    def __init__(self, name):
        self.name = name
        make_pie.cnt+=1
        
    @classmethod
    def number_Of_Pies(cls):
        print(f'파이를 {cls.cnt}명이 만들고 있습니다.')
        
    @classmethod
    def create(cls,name): # 클래스 내부에서 클래스 안 메소드를 호출하는 함수
        p=cls(name)
        return p
        
        
        
a = make_pie('Kevin')
b = make_pie('jane')
c = make_pie('john')

make_pie.number_Of_Pies()

make_pie.create('tom')
make_pie.create('bob')
make_pie.number_Of_Pies()
# @classmethod 라는 데코레이터를 사용함
# 호출시 첫번째 인자로 항상 cls를 사용하고 [cls는 class인 make_pie 자체 (self느낌)]
# cls는 클래스 자체를 의미하며 바로 make_pies를 뜻함
```
